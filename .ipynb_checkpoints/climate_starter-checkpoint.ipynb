```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import scipy.stats as sts
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# ispect measurement table
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# ispect station table
inspector = inspect(engine)
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent_date = session.query(measurement.date).order_by(measurement.date.desc()).first()
most_recent_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
year_age = dt.date(2017, 8, 23) - dt.date(2016, 8, 23)
query_date = dt.date(2017, 8, 23) - year_age


# Perform a query to retrieve the date and precipitation scores
results = session.query(measurement.date, measurement.prcp).\
    filter(measurement.date >= query_date).\
    order_by(measurement.date).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
precip_df = pd.DataFrame(results, columns = ['Date', 'Precipitation'])


# Sort the dataframe by date
precip_df.sort_values(by=['Date'])

# Convert Date column to Datetime
precip_df['Date'] = pd.to_datetime(precip_df['Date'])
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(precip_df['Date'].values,precip_df['Precipitation'].values, 
        label='precipitation')

#  Add labels
ax.set(xlabel='Date', ylabel='Inches')

# Format the x axis
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(DateFormatter("%b-%d-%y"))
ax.set_xlim(precip_df['Date'].min(),precip_df['Date'].max())
plt.xticks(rotation = 45)
plt.legend(loc='upper right')
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
std = precip_df['Precipitation'].std()
mean = precip_df['Precipitation'].mean()
min_p = precip_df['Precipitation'].min()
max_p = precip_df['Precipitation'].max()


quartiles = precip_df['Precipitation'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The minimum precipitation is: {min_p} inches")
print(f"The lower quartile of precipitation is: {lowerq} inches")
print(f"The interquartile range of precipitation is: {iqr}")
print(f"The median of precipitation is: {quartiles[0.5]} inches")
print(f"The upper quartile of precipitation is: {upperq} inches")
print(f"The maximum precipitation is: {max_p} inches")
print("")
print(f"The mean of the precipication data is {mean} inches")
print(f"The standard deviation of the precipication data is {std}")
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
station_count = session.query(station.id).count()
station_count
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
sel = [measurement.station,func.count(measurement.station)]
station_counts = session.query(*sel).\
    group_by(measurement.station).all()
station_counts
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
frequent_station = 'USC00519281'

# Calculate average temp
avg_temp_query = session.query(func.avg(measurement.tobs)).\
    filter(measurement.station == frequent_station).all()
avg_temp = avg_temp_query[0][0]

# Calculate highest temp
max_temp_query = session.query(func.max(measurement.tobs)).\
    filter(measurement.station == frequent_station).all()
max_temp = max_temp_query[0][0]

# Calculate lowest temp
min_temp_query = session.query(func.min(measurement.tobs)).\
    filter(measurement.station == frequent_station).all()
min_temp = min_temp_query[0][0]

print(f"The minimum temperature at station {frequent_station} was {min_temp}")
print(f"The maximum temperature at station {frequent_station} was {max_temp}")
print(f"The average temperature at station {frequent_station} was {avg_temp}")
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station
temp_data = session.query(measurement.tobs).filter(measurement.date >= query_date).\
    filter(measurement.station == frequent_station).\
    order_by(measurement.date).all()

# convert results to list instead of list of tuples
temp_data_list = []
for i in temp_data:
    temp_data_list.append(i[0])
```

```python
# plot the results as a histogram
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(temp_data_list, bins=12, alpha=0.75, label='tobs')

#  Add labels
ax.set(xlabel='Temperature', ylabel='Frequency')
ax.grid(True)
plt.legend(loc='upper right')
plt.show()
```
